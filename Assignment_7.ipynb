```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('world_population.csv')
print(df)
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Population'], marker='o', linestyle='-', color='green')
plt.xlabel('Year')
plt.ylabel('Population (in millions)')
plt.title('World Population Growth Over Years')
plt.grid(True)
plt.show()
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Read the CSV file
df = pd.read_csv('AAPL.csv')

# Convert 'Date' to datetime, coercing errors to NaT
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Drop rows with NaT in the 'Date' column (optional, depending on desired behavior)
df.dropna(subset=['Date'], inplace=True)

# Sort by date (if not already)
df = df.sort_values('Date')

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], color='blue', label='AAPL Close Price')

# Formatting the x-axis for dates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d')) # Changed format to include day
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))  # every 3 months

plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.title('Apple Inc. (AAPL) Stock Price Trend')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
