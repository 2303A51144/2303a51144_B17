```python
import pandas as pd

students = pd.DataFrame({
    'student_id': [11, 22, 33, 44, 55],
    'name': ['Sanskar', 'Vishwa', 'Charlie', 'Rohit', 'suu'],
    'gender': ['M', 'M', 'M', 'M', 'F']
})

exam_scores = pd.DataFrame({
    'student_id': [11, 22, 33, 66],
    'math_score': [98, 90, 78, 88],
    'english_score': [92, 85, 80, 75]
})

print("ðŸ“˜ Students Dataset:")
print(students)
print("\nðŸ“— Exam Scores Dataset:")
print(exam_scores)

inner_merge = pd.merge(students, exam_scores, on='student_id', how='inner')
print("\nðŸ”— Inner Join Result:")
print(inner_merge)

left_merge = pd.merge(students, exam_scores, on='student_id', how='left')
print("\nðŸ”— Left Join Result:")
print(left_merge)

right_merge = pd.merge(students, exam_scores, on='student_id', how='right')
print("\nðŸ”— Right Join Result:")
print(right_merge)

print("\nðŸ“Š Aggregate Statistics (Inner Join Data):")
print("Mean Scores:\n", inner_merge[['math_score', 'english_score']].mean())
print("Max Scores:\n", inner_merge[['math_score', 'english_score']].max())
print("Min Scores:\n", inner_merge[['math_score', 'english_score']].min())
print("Count of Scores:\n", inner_merge[['math_score', 'english_score']].count())

grouped_stats = inner_merge.groupby('gender')[['math_score', 'english_score']].agg(['mean', 'max', 'min', 'count'])
print("\nðŸ“Š Grouped Statistics by Gender:")
print(grouped_stats)
```

```python
import pandas as pd

customers = pd.read_csv("customers.csv")
sales = pd.read_csv("sales.csv")

print("Customers dataset shape:", customers.shape)
print("Sales dataset shape:", sales.shape)
print("\nCustomers columns:", customers.columns.tolist())
print("Sales columns:", sales.columns.tolist())
print("\nCustomers sample:\n", customers.head())
print("\nSales sample:\n", sales.head())

merged_inner = pd.merge(customers, sales, left_on='customer_id', right_on='customer_id', how='inner')
merged_left = pd.merge(customers, sales, left_on='customer_id', right_on='customer_id', how='left')

print("Inner Join Merge:\n", merged_inner.head(), "\n")
print("Left Join Merge:\n", merged_left.head(), "\n")

total_sales = sales['amount'].sum()
mean_sales = sales['amount'].mean()
sales_per_cust = sales.groupby('customer_id')['amount'].agg(['sum', 'mean', 'count']).reset_index()

print(f"Total sales: {total_sales:.2f}")
print(f"Average sale: {mean_sales:.2f}")
print("Sales per customer:\n", sales_per_cust.head(), "\n")

sales_with_country = pd.merge(sales, customers[['customer_id', 'region']], left_on='customer_id', right_on='customer_id', how='left')
region_stats = sales_with_country.groupby('region')['amount'].agg(['sum', 'mean', 'count']).sort_values(by='sum', ascending=False).reset_index()

print("Sales by region (Country):\n", region_stats)

best_region = region_stats.loc[region_stats['sum'].idxmax()]
print(f"Top performing region: {best_region['region']} with total sales {best_region['sum']:.2f}")

display(merged_inner)
```
